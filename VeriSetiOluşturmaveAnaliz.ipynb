```python
import pandas as pd
import matplotlib.pyplot as plt

veriSeti = pd.read_csv('/content/gym_members_exercise_tracking.csv')
YapilanEgzersizler = veriSeti['Workout_Type']
HarcananKalori = veriSeti['Calories_Burned']

plt.figure(figsize=(5,5))  # Grafiğin boyutunu ayarlayın
plt.bar(YapilanEgzersizler, HarcananKalori)
plt.xlabel('Yapılan egzersizler')
plt.ylabel('Harcanan Kalori')
plt.title('Spor Salonu Üyeleri Egzersiz İlerlemesi')
plt.xticks(rotation=90)  # Ülke isimlerinin dikey yazılması
plt.show()
```
